```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob
```

```python
df = pd.read_csv("IMDB Dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
def get_sentiment(text):

    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"

    elif polarity < 0:
        return "Negative"

    else:
        return "Neutral"
```

```python
df["Predicted_Sentiment"] = df["review"].apply(get_sentiment)

df.head()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(data=df,
              x="Predicted_Sentiment")

plt.title("Sentiment Distribution")

plt.show()
```

```python
df["Predicted_Sentiment"].value_counts().plot.pie(
    autopct="%1.1f%%"
)

plt.title("Percentage of Sentiments")

plt.ylabel("")

plt.show()
```

Positive reviews dominate the dataset.

Negative reviews form a smaller proportion.

Neutral reviews are comparatively fewer.

The overall customer opinion is largely positive.

This project analyzed movie reviews using TextBlob sentiment analysis.

Each review was classified as Positive, Negative, or Neutral.
